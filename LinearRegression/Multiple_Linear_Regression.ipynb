```python
import pandas as pd

# For data manipulation
import matplotlib.pyplot as plt
import seaborn as sns

#For building the model and evaluating it
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import numpy as np

#set a style for our plots
sns.set(style="whitegrid")

df = pd.read_csv("advertising.csv")
# print(df.head())
sns.pairplot(df)
plt.show()

x= df[["TV","Radio","Newspaper"]]
print(x)
y = df["Sales"]
print(y)
```

```python
# Split the data into training and testing sets
# 80% for training and 20% for testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print("Training set shape (x.train)", x_train.shape)
print("Training set shape (x.test)", x_test.shape)

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(x_train, y_train)

# print the coefficients for each feature
print("TV: ",model.coef_[0])
print("Radio:",model.coef_[1])
print("Newspaper:",model.coef_[2])

# print the intercept
print("intercept: ",model.intercept_)

# Predict the sales using the test set
y_pred_simple = model.predict(x_test)

#Calculate  and print the evaluation metrics
mse_value = metrics.mean_squared_error(y_test, y_pred_simple)
print(f"Mean Squared Error (MSE): {mse_value:.4f}")
print(type(mse_value))

#Find the Root Mean Squared Error (RMSE)
rmse_value = np.sqrt(mse_value)
print(f"Root Mean Squared Error (RMSE): {rmse_value:.4f}")
```
